```python
#  train and test
import pandas as pd
from spacy import *
from nltk import *
import numpy as np
from scipy import stats
from scipy.stats import norm,expon,binom,poisson
from collections import Counter
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
from numpy import random
from sklearn.metrics import r2_score
```

```python
np.random.seed(2)
page = np.random.normal(3.0,1.0,100)
amount = np.random.normal(50.0,50.0,100)/page
sns.scatterplot(page,amount)
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(page,amount)
```

```python
trainX = page[:80]
testx = page[80:]

trainY = amount[:80]
testy = amount[80:]
```

```python
plt.scatter(trainX,trainY)
```

```python
plt.scatter(testx,testy)
```

```python
x = np.array(trainX)
y = np.array(trainY)

p4 = np.poly1d(np.polyfit(x,y,8))
xp = np.linspace(0,7,100)
axes = plt.axes()
axes.set_xlim(0,7)
axes.set_ylim(0,200)
plt.scatter(x,y)
plt.plot(xp,p4(xp),c='r')
```

```python
testx = np.array(testx)
testy = np.array(testy)

axes = plt.axes()
axes.set_xlim(0,7)
axes.set_ylim(0,200)
plt.scatter(testx,testy)
plt.plot(xp,p4(xp),c='r')
```

```python
r2_score(testy,p4(testx))
```

```python
r2_score(np.array(trainY),p4(np.array(trainX)))
```
